```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('states_all.csv')
df.describe()
```

```python
yıllar=df['YEAR'].unique()
def doldur(df):
       for sutun_adi in df.columns[3:]:
            for yıl in yıllar:
                df[sutun_adi].fillna(df.groupby(df['YEAR']==yıl)[sutun_adi].transform('mean'), inplace=True)

doldur(df)
df
```

```python
df = pd.read_csv('states_all.csv')
```

```python
MATH_4 = df.AVG_MATH_4_SCORE.dropna()
READING_4 = df.AVG_READING_4_SCORE.dropna()
MATH_8 = df.AVG_MATH_8_SCORE.dropna()
READING_8 = df.AVG_READING_8_SCORE.dropna()
```

ortalama ile dolduralım

```python
mean_MATH_4 = df.AVG_MATH_4_SCORE.fillna(df.AVG_MATH_4_SCORE.mean())
mean_READING_4 = df.AVG_READING_4_SCORE.fillna(df.AVG_READING_4_SCORE.mean())
mean_MATH_8 = df.AVG_MATH_8_SCORE.fillna(df.AVG_MATH_8_SCORE.mean())
mean_READING_8 = df.AVG_READING_8_SCORE.fillna(df.AVG_READING_8_SCORE.mean())
```

medyan ile dolduralım

```python
median_MATH_4 = df.AVG_MATH_4_SCORE.fillna(df.AVG_MATH_4_SCORE.median())
median_READING_4 = df.AVG_READING_4_SCORE.fillna(df.AVG_READING_4_SCORE.median())
median_MATH_8 = df.AVG_MATH_8_SCORE.fillna(df.AVG_MATH_8_SCORE.median())
median_READING_8 = df.AVG_READING_8_SCORE.fillna(df.AVG_READING_8_SCORE.median())
```

enterpolasyon ile dolduralım

```python
interpolate_MATH_4 = df.AVG_MATH_4_SCORE.interpolate()
interpolate_READING_4 = df.AVG_READING_4_SCORE.interpolate()
interpolate_MATH_8 = df.AVG_MATH_8_SCORE.interpolate()
interpolate_READING_8 = df.AVG_READING_8_SCORE.interpolate()
```

```python
plt.figure(figsize=(20,20))
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }

plt.subplot(4,4,1)
plt.hist(MATH_4)
plt.title("AVG_MATH_4_SCORE",fontdict=baslik_font)

plt.subplot(4,4,2)
plt.hist(mean_MATH_4)
plt.title("AVG_MATH_4_SCORE (mean)",fontdict=baslik_font)

plt.subplot(4,4,3)
plt.hist(median_MATH_4)
plt.title("AVG_MATH_4_SCORE (median)",fontdict=baslik_font)

plt.subplot(4,4,4)
plt.hist(interpolate_MATH_4)
plt.title("AVG_MATH_4_SCORE (interpolated)",fontdict=baslik_font)

plt.subplot(4,4,5)
plt.hist(READING_4)
plt.title("AVG_MATH_4_SCORE",fontdict=baslik_font)

plt.subplot(4,4,6)
plt.hist(mean_READING_4)
plt.title("AVG_MATH_4_SCORE (mean)",fontdict=baslik_font)

plt.subplot(4,4,7)
plt.hist(median_READING_4)
plt.title("AVG_MATH_4_SCORE (median)",fontdict=baslik_font)

plt.subplot(4,4,8)
plt.hist(interpolate_READING_4)
plt.title("AVG_MATH_4_SCORE (interpolated)",fontdict=baslik_font)

plt.subplot(4,4,9)
plt.hist(MATH_8)
plt.title("AVG_MATH_4_SCORE",fontdict=baslik_font)

plt.subplot(4,4,10)
plt.hist(mean_MATH_8)
plt.title("AVG_MATH_4_SCORE (mean)",fontdict=baslik_font)

plt.subplot(4,4,11)
plt.hist(median_MATH_8)
plt.title("AVG_MATH_4_SCORE (median)",fontdict=baslik_font)

plt.subplot(4,4,12)
plt.hist(interpolate_MATH_8)
plt.title("AVG_MATH_4_SCORE (interpolated)",fontdict=baslik_font)

plt.subplot(4,4,13)
plt.hist(READING_8)
plt.title("AVG_READING_8_SCORE",fontdict=baslik_font)

plt.subplot(4,4,14)
plt.hist(mean_READING_8)
plt.title("AVG_READING_8_SCORE (mean)",fontdict=baslik_font)

plt.subplot(4,4,15)
plt.hist(median_READING_8)
plt.title("AVG_READING_8_SCORE (median)",fontdict=baslik_font)

plt.subplot(4,4,16)
plt.hist(interpolate_READING_8)
plt.title("AVG_READING_8_SCORE (interpolated)",fontdict=baslik_font)


plt.tight_layout()
plt.show()
```
