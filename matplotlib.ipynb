```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('./movie.csv')
df
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.bar(df['Year'],df['Rating'], color='pink', width=.5)
plt.xlabel('yıl')
plt.ylabel('imdb')
plt.title("yıl-imdb")

plt.subplot(1,2,2)
plt.bar(df['Year'],df['Runtime (Minutes)'], color='grey', width=.5)
plt.xlabel('yıl')
plt.ylabel('süre')
plt.title("yıl-süre")

plt.show()
```
