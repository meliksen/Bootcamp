```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.figure(figsize=(10,4))

dagilim = np.random.normal(5,10,100)

plt.hist(dagilim) 
plt.show()
```

```python
standart_normal=np.random.normal(0,1,100)

standart_normal.sort()
dagilim.sort()

plt.figure(figsize=(10,4), dpi = 200)
plt.scatter(standart_normal,dagilim)
plt.plot([np.min(standart_normal), np.max(standart_normal)],
         [np.min(dagilim), np.max(dagilim)], c='red')

plt.title("Q-Q izlm")
plt.ylabel("Normal Dalm")
plt.xlabel("Standart Normal Dalm")

plt.show()
```
