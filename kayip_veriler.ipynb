```python
def null_degerler(df):
    null_df=pd.DataFrame()
    null_df['yüzde']=df.isnull().sum()*100/df.shape[0]
    null_df['adet']=df.isnull().sum()
    return null_df.sort_values('yüzde',ascending=False)
null_degerler(df)
```
